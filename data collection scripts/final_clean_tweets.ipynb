```python
import re
import pandas as pd
import re

d_f = pd.read_csv("C:/Users/dues1/Desktop/DIC2/tweets_trump_1_filter_oneday.csv")
string = ""
for i in range(len(d_f)):
    string = string + "charanya" + d_f.loc[i, 'text'] 

clean_string = re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', string)
clean_string1 = re.sub(r"[^a-zA-Z]+", ' ', clean_string)
clean_string1 = ' '.join([w for w in clean_string1.split() if len(w)>1])
clean_string1 = clean_string1.lower()
clean_string1 = re.sub("twitter", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("think", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("would", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("thehill", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("going", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("april", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("could", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("still", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("wants", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("before", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("there", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("every", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("never", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("someone", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("believe", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("actually", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("stormy", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("really", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("month", ' ', clean_string1, flags=re.I)
clean_string1 = re.sub("behind", ' ', clean_string1, flags=re.I)

with open("tweets_newline_mapper_input_oneday.txt", "w") as text_file:
    text_file.write("Tweets : %s" % clean_string1)
```

```python
# Read in the file
with open('C:/Users/dues1/Desktop/tweets_newline_mapper_input_oneday.txt', 'r') as file :
    filedata = file.read()

# Replace the target string
filedata = filedata.replace('charanya', '\n')

# Write the file out again
with open('C:/Users/dues1/Desktop/tweets_newline_mapper_input_1dayreplaced.txt', 'w') as file:
    file.write(filedata)
```
