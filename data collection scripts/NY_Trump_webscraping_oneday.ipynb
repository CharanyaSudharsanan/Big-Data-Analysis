```python
from nytimesarticle import articleAPI
api = articleAPI('8b18491cab36479486581b2c19309a5e')
articles = api.search( q = "trump", begin_date = 20180323, end_date = 20180324)
print(articles)
```

```python
import math
hits = articles['response']['meta']['hits']
print("hits :" + str(hits))
ls = []
pages = int(math.ceil(hits/10))
for i in range(pages):
    print("collecting page " + str(i))
    articles['page'] = i
    for article in articles['response']['docs']:
        ar = article['web_url'].encode('utf-8')
        if not ar.__contains__("video"):
            ls.append(article['web_url'].encode('utf-8'))
        else:
            continue
print(ls)
        

    
```

```python
from bs4 import BeautifulSoup
import requests
import unicodedata
artc = ""

for x in ls[:]:
    page = requests.get(x)
    soup = BeautifulSoup(page.content, 'html.parser')
    soup = soup.find('article')
    for script in soup(["script", "style"]):
        script.extract() 
    text = soup.text.strip()
    lines = (line.strip() for line in text.splitlines())
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    text = ' '.join(chunk for chunk in chunks if chunk)
    artc1 = text.encode('utf-8').strip()
    artc = artc + "charanya" + artc1
```

```python
import re
clean_a = re.sub(r"[^a-zA-Z]+", ' ', artc)
clean_a = ' '.join([w for w in clean_a.split() if len(w)>1])
clean_a = clean_a.lower()
clean_a = re.sub("New York Times", ' ', clean_a, flags=re.I)
clean_a = re.sub("story", ' ', clean_a, flags=re.I)
clean_a = re.sub("reading", ' ', clean_a, flags=re.I)
clean_a = re.sub("continue", ' ', clean_a, flags=re.I)
clean_a = re.sub("house", ' ', clean_a, flags=re.I)
clean_a = re.sub("times", ' ', clean_a, flags=re.I)
clean_a = re.sub("special", ' ', clean_a, flags=re.I)
clean_a = re.sub("would", ' ', clean_a, flags=re.I)
clean_a = re.sub("could", ' ', clean_a, flags=re.I)
clean_a = re.sub("photo", ' ', clean_a, flags=re.I)
clean_a = re.sub("today", ' ', clean_a, flags=re.I)
clean_a = re.sub("appears", ' ', clean_a, flags=re.I)
with open("trump_articles_clean_oneday.txt", "w ") as text_file:
    text_file.write("NYTimes %s" % clean_a)
```
